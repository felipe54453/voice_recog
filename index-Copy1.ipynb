## Transcrição

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
MODEL_NAME = "facebook/wav2vec2-base-960h"

# o large gera OOM só de carregar o modelo
#MODEL_NAME = "facebook/wav2vec2-large-xlsr-53-portuguese"
```

```python
from transformers import Wav2Vec2Tokenizer, Wav2Vec2ForCTC
import gc

def transcribe_file(model_name, file, offset_seconds=0, max_seconds=2):
    # carregar áudio
    speech, sr = torchaudio.load(file)
    duration = speech.shape[1] / sr
    
    if offset_seconds >= duration:
        print(f"Áudio possui {duration:.2f} segundos, mas offset especificado é de {offset_seconds}.")
        return
    
    # crop em max_seconds (primeiro eixo são os canais do áudio)
    speech = speech[:, offset_seconds*sr:offset_seconds*sr+max_seconds*sr]
    
    
    # resampling para 16.000 kHz
    resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16_000)
    speech = resampler.forward(speech.squeeze(0)).numpy()
    del resampler
    gc.collect()

    # tokenize
    tokenizer = Wav2Vec2Tokenizer.from_pretrained(model_name)
    input_values = tokenizer(speech, return_tensors="pt", padding="longest").input_values  # Batch size 1
    del speech, tokenizer
    gc.collect()

    # retrieve logits
    model = Wav2Vec2ForCTC.from_pretrained(model_name)
    with torch.no_grad():
        logits = model(input_values).logits
    del model, input_values
    gc.collect()

    # take argmax and decode
    predicted_ids = torch.argmax(logits, dim=-1)
    del logits
    gc.collect()

    tokenizer = Wav2Vec2Tokenizer.from_pretrained(model_name)
    transcription = tokenizer.batch_decode(predicted_ids)
    del tokenizer
    gc.collect()
        
    return transcription
```
